> Tomaremos estos datos como datos erróneos ya que no existe una propiedad de una categoría mayor o igual a dos. Igualaremos todos datos en las columnas Cat_X mayores o iguales a dos, a cero.


```python
def handle_change(cell):
    if cell > 1 or cell < 0:
        return 0
    else:
        return cell
```

```python
num_iterations = 515  # Num de cats máximo existente en training_data

for i in range(1, num_iterations):
    column_name = f"Cat_{i}"

    if column_name in training_data.columns:
        for j in range(0, len(training_data[column_name])):
            training_data.loc[j, column_name] = handle_change(
                training_data.loc[j, column_name]
            )  # Reemplazando datos
    else:
        print(f"La columna {column_name} no existe en el conjunto de datos.")
```

```python
print(training_data["Cat_1"].head())
print(training_data["Cat_2"].head())
print(training_data["Cat_3"].head())
```

#### <strong>Hallando y eliminando posibles columnas irrelevantes en Cat_x</strong>

> Consideramos columnas irrelevantes las columnas en las cuales todos sus valores para la categoría son iguales.


```python
count = 0

for i in range(1, num_iterations):
    column_name = f"Cat_{i}"

    if column_name in training_data.columns:
        min_val = min(training_data[column_name])
        max_val = max(training_data[column_name])

        if min_val == max_val:
            if min_val == 0 or min_val == 1:
                count += 1
                print(
                    column_name, "eliminada"
                )  # Columnas en las cuales todos sus valores son 0 o 1
                training_data = training_data.drop(
                    column_name, axis=1
                )  # Eliminando columna
    else:
        print(f"La columna {column_name} no existe en el conjunto de datos.")

print(f"Total de cats con todos sus campos en 0 o 1: {count}")
```

#### <strong>Buscando datos nulos en el data frame</strong>


```python
# Halla columnas con datos nulos
def look_for_nullData(df):
    columns_with_null_data = df.columns[df.isnull().any()]

    columns_with_null_data.isnull().sum()

    for col in columns_with_null_data:
        null_count = df[col].isnull().sum()
        print(col, ":", null_count)
```

```python
look_for_nullData(training_data)
```

#### <strong>Limpiando datos nulos en columnas Quan_X</strong>

> Las columnas Quan_X son columnas cuantitativas. No encontramos algún tipo de patron en los datos nulos el cual nos pueda dar una razón coherente del porque los datos son nulos, por ende y para evitar alteraciones significativas en las métricas de las columnas, rellenaremos estos espacios con la media de la columna.


```python
num_iterations = 31  # Num de quans máximo existente en training_data

for i in range(1, num_iterations):
    column_name = f"Quan_{i}"

    if column_name in training_data.columns:
        mean_column = training_data[column_name].mean()

        for j in range(0, len(training_data[column_name])):
            if pd.isnull(training_data.loc[j, column_name]):
                training_data.loc[j, column_name] = mean_column  # Reemplazando datos
    else:
        print(f"La columna {column_name} no existe en el conjunto de datos.")
```

```python
look_for_nullData(training_data)
```

> Encontramos un error al nombrar algunos Quan que fueron nombrados como Quant. Quant no aparece como algunas de las variables entregadas por el cliente. Coincide que las Quan inexistentes (23, 24 y 25) existen en Quant a excepción de Quan 22 y Quant 22 debido a que ambas existen. En ese caso tomaremos la que mas datos tiene (Quant_22).


```python
print(training_data["Quan_22"].describe())
print(training_data["Quant_22"].describe())
```

```python
# Reemplaza los datos null de los Quant por su respectiva media y renombra cada Quant_X por Quan_X
def clean_and_rename(num, training_data):
    column = f"Quant_{num}"
    mean = training_data[column].mean()

    for i in range(0, len(training_data[column])):
        if pd.isnull(training_data.loc[i, column]):
            training_data.loc[i, column] = mean  # Reemplazando datos

    training_data.rename(columns={column_name: f"Quan_{num}"}, inplace=True)
    training_data = training_data.drop(column, axis=1)
```

```python
training_data = training_data.drop("Quan_22", axis=1)  # Eliminando columna Quan_22

clean_and_rename(22, training_data=training_data)
clean_and_rename(23, training_data=training_data)
clean_and_rename(24, training_data=training_data)
clean_and_rename(25, training_data=training_data)
```

```python
look_for_nullData(df=training_data)
```

#### <strong>Limpiando datos nulos en columnas Date_2 y Outcome_MX</strong>

> En el caso de Date_2, podemos notar a continuación que siempre las fechas Date_1 son mayores a las correspondientes en Date_2.


```python
print(training_data["Date_1"])
print(training_data["Date_2"])

for i in range(0, len(training_data["Date_1"])):
    if training_data["Date_1"][i] < training_data["Date_2"][i]:
        print("Existe.")  # Si existe un Date_1 menor a Date_2, se imprime Existe.
```

```python
print(training_data["Date_1"].describe())
print(training_data["Date_2"].describe())
```

> Debido a que el menor de Date_2, nunca es mayor al menor de Date_1 entonces aquellos datos nulos en Date_2 serán iguales al menor de Date_2. De esta manera evitamos hacerlo con la media, lo cual podría arrojar incongruencias (Por ejemplo, con un Date_2 > Date_1) o 0 lo cual tampoco tiene sentido con la descripción de los campos Date_2.


```python
column = training_data["Date_2"]
min_val = min(column)

for i in range(0, len(column)):
    if pd.isnull(training_data.loc[i, "Date_2"]):
        training_data.loc[i, "Date_2"] = min_val  # Reemplazando datos
```

> Para los Outcomes_MX que son ventas mensuales, para no afectar las métricas de estos datos, implementaremos la media de las ventas.


```python
num_iterations = 13  # Num de Outcomes máximo existente en training_data

for i in range(1, num_iterations):
    column_name = f"Outcome_M{i}"

    if column_name in training_data.columns:
        mean_column = training_data[column_name].mean()

        for j in range(0, len(training_data[column_name])):
            if pd.isnull(training_data.loc[j, column_name]):
                training_data.loc[j, column_name] = mean_column  # Reemplazando datos
    else:
        print(f"La columna {column_name} no existe en el conjunto de datos.")
```
